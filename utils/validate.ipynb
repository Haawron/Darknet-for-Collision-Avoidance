```python
import glob, csv, cv2, random
import matplotlib.pyplot as plt

def imshow(img):
    plt.figure(figsize=(15, 10))
    plt.axis('off')
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show() 
```

```python
tmp = random.choice(glob.glob('data/DBPAS_train3/*_c*'))[:-4]
print(tmp)
with open(tmp+'.txt', 'r') as f:
    data = list(map(float, next(csv.reader(f, delimiter=' ')))) 
sx, sy = map(int, tmp[-7:].split('x'))
cx, cy = int(sx*float(data[1])), int(sy*float(data[2]))
bx, by = int(sx*float(data[3])), int(sy*float(data[4]))
boxed = cv2.rectangle(
    cv2.imread(tmp+'.jpg'),
    (cx-bx//2, cy-by//2),
    (cx+bx//2, cy+by//2),
    (0, 0, 255), 1) 
imshow(boxed)
```
